```python
# create dataset 
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, centers=5, random_state=15)
```

```python
X.shape
```

```python
# create scatter plot for samples from each class
for class_value in range(5):
    # get row indexes for samples with this class
    row_ix = where(y == class_value)
    # create scatter of these samples
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
pyplot.show()
```

```python
def test_clustering (alg, X, y) :
    # define the model
    model = alg(cluster_all=False)
    # fit model and predict clusters
    yhat = model.fit_predict(X)
    # retrieve unique clusters
    clusters = unique(yhat)
    # create scatter plot for samples from each cluster
    for cluster in clusters:
    # get row indexes for samples with this cluster
        row_ix = where(yhat == cluster)
    # create scatter of these samples
        pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
    # show the plot
    pyplot.show()
    return yhat
```

```python
def parse_labels(yhat):
    cluster_map = pd.DataFrame()
    for i in range(len(yhat)):
        cluster_map['data_index'] = i
    return cluster_map[cluster_map.cluster == 3]
```

```python
def ClusterIndices(clustNum, labels_array): 
    return np.where(labels_array == clustNum)[0]
```

```python
def find_means(folders):
    means = []
    for i in range(len(folders)):
        means.append(np.mean(folders[i]))
    return means
```
